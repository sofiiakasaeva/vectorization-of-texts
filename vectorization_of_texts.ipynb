```python
# load files
import os

filenames = [
    "text1_for_vectorization.txt",
    "text2_for_vectorization.txt",
    "text3_for_vectorization.txt",
    "text4_for_vectorization.txt"
]

labels = [
    "Text 1: Limerence: When a crush becomes obsessive",
    "Text 2: A bike ride through Patagonia",
    "Text 3: Why we keep dating people like our ex",
    "Text 4: Changi Airport"
]

texts = []
for fname in filenames:
    with open(fname, encoding='utf-8') as f:
        text = f.read()
        texts.append(text)
        print(f'{fname}: {len(text)} characters')
```

```python
from sklearn.feature_extraction.text import TfidfVectorizer

vectorizer = TfidfVectorizer(
    stop_words="english", # удаляем популярные английские слова
    lowercase=True,
    ngram_range=(1, 1)
)

tfidf_matrix = vectorizer.fit_transform(texts)
```

```python
from sklearn.metrics.pairwise import cosine_similarity
import pandas as pd

cos_sim = cosine_similarity(tfidf_matrix)

sim_df = pd.DataFrame(cos_sim, index=labels, columns=labels)
print("Cosine Similarity Matrix:")
print(sim_df.round(4))
```
